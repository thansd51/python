```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston
boston = load_boston()
print(boston)
```

# 2) 데이터 수집, 준비 및 탐색

```python
print(boston.DESCR)
```

```python
boston_df = pd.DataFrame(boston.data, columns=boston.feature_names)
boston_df.head()
```

```python
boston_df['PRICE'] = boston.target
boston_df.head()
```

```python
print('보스턴 주택 가격 데이터셋 크기:', boston_df.shape)
```

```python
boston_df.info()
```

# 3) 분석 모델 구축, 결과 분석 및 시각화

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# X, Y 분할
Y = boston_df['PRICE']
X = boston_df.drop(columns=['PRICE'])
```

```python
# 훈련용 데이터와 평가용 데이터 분할
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=156)
```

```python
# 선형 회귀 분석 : 모델 생성
Ir = LinearRegression()
```

```python
# 선형 회귀 분석: 모델 훈련
Ir.fit(X_train, Y_train)
```

```python
# 선형 회귀 분석 : 평가 데이터에 대한 예측 수행 -> 예측 결과 Y_predict 구하기
Y_predict = Ir.predict(X_test)
```

# 4) 회귀 분석 결과를 산점도 + 선형 회귀 그래프로 시각화하기

```python
mse = mean_squared_error(Y_test, Y_predict)
rmse = np.sqrt(mse)
print('MSE : {0:.3f}, RMSE : {1:.3f}'.format(mse, rmse))
print('R^2(Variance score) : {0:.3f}'.format(r2_score(Y_test, Y_predict)))
```

```python
print('Y 절편 값 :', Ir.intercept_)
print('회귀 계수 값 :', np.round(Ir.coef_, 1))
```

```python
coef = pd.Series(data=np.round(Ir.coef_, 2), index=X.columns)
coef.sort_values(ascending=False)
```
