# 자동차 연비 예측 분석 ==> 회귀분석

```python
# numpy, pandas 를 import
# auto-mpg.csv 파일을 읽어 data_df 에 넣기
import numpy as np
import pandas as pd

data_df = pd.read_csv('auto-mpg.csv', header=0)
data_df
```

```python
# 데이터 셋 크기 확인
data_df.shape
```

```python
# data_df에서 위에서 5개 데이터 확인
data_df.head()
```

```python
# data_df에 데이터 유형 확인
data_df.info()
```

```python
# 'car_name', 'origin', 'horsepower' 제거
data_df = data_df.drop(columns=['car_name', 'origin', 'horsepower'])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# x(mpg 제외환 나머지), Y(mpg) 분할
Y = data_df['mpg']
X = data_df.drop(columns=['mpg'])
X
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

```python
Y_predict = lr.predict(X_test)
Y_predict
```

```python
# 회귀 분석 결과를 산점도 + 선형 회귀 그래프로 시각화화기
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, axs = plt.subplots(figsize =(16, 16), ncols=3, nrows=2)
x_features = ['cylinders', 'displacement', 'weight', 'acceleration', 'model_year']
plot_color=['r', 'b', 'y', 'g', 'r']
for i, feature in enumerate(x_features):
    row = int(i/3)
    col = i%3
    sns.regplot(x=feature, y='mpg', data=data_df, ax=axs[row][col], color=plot_color[i])
```

```python
# 실제값과 예측값의 차이를 제곱으로 평균(MSE)
import numpy as np
# 실제값 : Y_test, 예측값 : Y_predict
mse = mean_squared_error(Y_test, Y_predict)
rmse = np.sqrt(mse)
# 결정계수(0 =< R2 =< 1) : 1에 가까울 수록 회귀식 정확도가 높다
print('mse: ',mse, 'rmse: ', rmse)
print('R^2: {0:.3f}'.format(r2_score(Y_test, Y_predict))) 
```

```python
print('Y절편 값: ', lr.intercept_)
print('회귀 계수(기울기) 값: ', np.round(lr.coef_, 2))
```

```python
coef = pd.Series(data=np.round(lr.coef_,2), index=X.columns)
coef.sort_values(ascending=False)
```
