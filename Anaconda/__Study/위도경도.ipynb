```python
# ex를 데이터프레임으로
ex = DataFrame(ex)
ex
```

```python
# 지도의 중심을 지정하기 위해 위도, 경도 평균구하기
lat = ex['위도'].mean()
long = ex['경도'].mean()
lat
```

```python
# 지도 띄우기
m = folium.Map([lat, long], zoom_start=9)
m
```

```python
for i in ex.index:
    sub_lat = ex.loc[i, '위도']
    sub_long = ex.loc[i, '경도']
    title = ex.loc[i, '구분']
    # 지도에 데이터 찍어서 보여주기
    folium.Marker([sub_lat, sub_long], tooltip=title).add_to(m)
m.save('example.html')
```
